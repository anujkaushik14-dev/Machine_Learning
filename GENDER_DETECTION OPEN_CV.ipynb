```python
import cv2
from deepface import DeepFace

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Start webcam
cap = cv2.VideoCapture(0)

frame_count = 0
gender_text = ""

while True:

    ret, frame = cap.read()

    if not ret:
        break

    # Resize frame for better speed
    frame = cv2.resize(frame, (640, 480))

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5,minSize=(30, 30))

    for (x, y, w, h) in faces:

        # Draw rectangle
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        # Run DeepFace every 30 frames
        if frame_count % 30 == 0:

            face_roi = frame[y:y+h, x:x+w]

            try:
                result = DeepFace.analyze(face_roi,actions=['gender'],enforce_detection=False,detector_backend='opencv')

                # Get gender result
                gender = result[0]['dominant_gender']

                gender_text = f"Gender: {gender}"

            except Exception as e:
                print("Error:", e)

        # Display text
        cv2.putText(frame,gender_text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

    frame_count += 1

    # Show window
    cv2.imshow("Gender Detection", frame)

    # ESC key to exit
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()

cv2.destroyAllWindows()
```
