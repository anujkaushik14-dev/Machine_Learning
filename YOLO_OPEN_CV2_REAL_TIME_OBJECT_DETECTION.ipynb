#### REAL-TIME DETECTION WITH YOLO

```python
from ultralytics import YOLO
import cv2

# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Open video
cap = cv2.VideoCapture(
    r"C:\Users\vv\OneDrive\Desktop\PYTHON\DNN\Old Delhi traffic flow _ Near Red Fort with Jain temple view.mp4"
)

if not cap.isOpened():
    raise IOError("Cannot open video")

while True:

    ret, frame = cap.read()

    if not ret:
        break

    # Resize frame
    frame = cv2.resize(frame, (640, 480))

    # Run YOLO detection
    results = model(frame)

    # Draw detections
    for result in results:

        boxes = result.boxes

        for box in boxes:

            # Confidence
            conf = float(box.conf[0])

            # Class ID
            cls = int(box.cls[0])

            # Class name
            class_name = model.names[cls]

            # Skip low confidence
            if conf < 0.5:
                continue

            # Coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # Draw rectangle
            cv2.rectangle(frame,
                          (x1, y1),
                          (x2, y2),
                          (0, 255, 0),
                          2)

            # Label
            label = f"{class_name} {conf:.2f}"

            cv2.putText(frame,
                        label,
                        (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.7,
                        (0, 255, 0),
                        2)

    # Show output
    cv2.imshow("YOLOv8 Detection", frame)

    # ESC to exit
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
