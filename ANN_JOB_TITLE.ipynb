```python
le_dict = {}

categorical_cols = ['job_title',
    'job_category','experience_level','education_required',
    'salary_tier','city','job_id','country',
    'company_size','industry','required_skills','remote_work'
]

for col in categorical_cols:
    le_dict[col] = LabelEncoder()
    df[col] = le_dict[col].fit_transform(df[col])
```

```python
x = df.drop("job_title", axis=1)
y = df.iloc[0:, 1:2]
```

```python
# Train-Test Split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
# Build ANN Model and HIDDEN layers
# Index 0 (train_sc.shape[0]): Represents the total number of rows (sample size).
# Index 1 (train_sc.shape[1]): Represents the total number of columns (features or variables).

classifier = Sequential()

classifier.add(Dense(units=128, activation='relu',input_shape=(x_train_sc.shape[1],)))
classifier.add(Dropout(0.2))
```

```python
classifier.add(Dense(units=64, activation='relu'))
classifier.add(Dropout(0.2))
```

```python
# classifier.add(Dense(units=32, activation='relu'))
```

```python
classifier.add(Dense(units=25,activation='softmax'))
```

```python
# Compile Model

classifier.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train Model

classifier.fit(x_train_sc, y_train, batch_size=32, epochs=100)
```

```python
def recomand(*features):
    input_data = np.array([features])
    scaled_input = sc.transform(input_data)

    pred = classifier.predict(scaled_input, verbose=0)

    pred_class_idx = np.argmax(pred)

    job_title = le_job.inverse_transform([pred_class_idx])[0]
    confidence = float(pred[0][pred_class_idx]) * 100

    return job_title, confidence
```
