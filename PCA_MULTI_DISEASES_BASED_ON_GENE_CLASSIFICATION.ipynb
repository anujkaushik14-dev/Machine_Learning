```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\vv\OneDrive\Desktop\PYTHON\Unsupervised\GeneX_3000_Synthetic_Genomic_Dataset.csv")
df
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder



label_encoders = {}
categorical_columns = ['Gender','Disease_Label']	

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df
```

```python
x=df.iloc[:, 1:801].values
x.shape
```

```python
y=df.iloc[:,3004].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)

x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train, y_train)
```

```python
pred=knn.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

results=confusion_matrix(y_test, pred)
print('confusion_matrix')
print(results)
print('Accuracy Score:', accuracy_score(y_test, pred))
```

```python
from sklearn.preprocessing  import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test-sc.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
x_train_new=pca.fit_transform(x_train)
x_test_new=pca.transform(x_test)
```

```python
print(x_train.shape)
```

```python
print(x_train_new.shape)
```

```python
pca.explained_variance_ratio_
```

```python
knn.fit(x_train_new, y_train)
```

```python
pred_new=knn.predict(x_test_new)
pred_new
```

```python
accuracy_score(y_test, pred_new)
```
