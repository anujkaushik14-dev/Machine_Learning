# CNN_PLANT_DISEASES_DETECTION

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np
```

```python
CNN_Classifier=Sequential()
```

```python
# step 1 : Convultion

CNN_Classifier.add(Conv2D(32,(3,3),input_shape=(64,64,3), activation='relu'))

# step:2 Pooling

CNN_Classifier.add(MaxPooling2D(pool_size=(2,2)))

# step:3 2nd Convulution

CNN_Classifier.add(Conv2D(32,(3,3), activation='relu'))

# step: Pooling

CNN_Classifier.add(MaxPooling2D(pool_size=(2,2)))

# step 4: Flatten

CNN_Classifier.add(Flatten())

# step 5: full connection

CNN_Classifier.add(Dense(units=32, activation='relu'))

CNN_Classifier.add(Dense(units=64, activation='relu'))

CNN_Classifier.add(Dense(units=128, activation='relu'))

CNN_Classifier.add(Dense(units=256, activation='relu'))

CNN_Classifier.add(Dense(units=256, activation='relu'))

CNN_Classifier.add(Dense(units=38, activation='softmax'))


# step 6: COMPILE ON

CNN_Classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# step :7 Fitting CNN to images

from keras_preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255, # to rescale the image in range of [0,1]
    rotation_range=20, 
    zoom_range=0.2, # to randomly zoom the images
    shear_range=0.2, #to randomly shear the images
    horizontal_flip=True # for randomly flipping half of the images horizontally
)

test_datagen= ImageDataGenerator(rescale= 1./255)
print(" \nTraining the data ..\n")
train_data_path=r"C:\Users\vv\Downloads\New folder\New Plant Diseases Dataset(Augmented)\New Plant Diseases Dataset(Augmented)\train"
test_Data_path=r"C:\Users\vv\Downloads\New folder\New Plant Diseases Dataset(Augmented)\New Plant Diseases Dataset(Augmented)\test"

training_set=train_datagen.flow_from_directory( train_data_path,target_size=(64,64),
                                               batch_size=12, class_mode='categorical')

test_set = test_datagen.flow_from_directory(test_Data_path,target_size=(64,64),
                                            batch_size=16, class_mode='categorical',shuffle=False)

```

```python
history=CNN_Classifier.fit(training_set,epochs=5, verbose=1,  validation_data=test_set)

CNN_Classifier.save('plant_diseases_model.h5')
print("Model Saved")
```
