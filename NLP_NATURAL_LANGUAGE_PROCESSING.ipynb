### GET DATA IN DATA FRAME WITH REGULAR EXPRESSION USE

```python
import re
from ddgs import DDGS
from collections import defaultdict
import urllib.request
from bs4 import BeautifulSoup


movie = input("Enter Movie Name: ")
query = movie + " user review"

t=[]

with DDGS() as ddgs:
    results = ddgs.text(query, max_results=5)

    for r in results:
        # print(f"Title: {result['title']}")
        # print(r["href"])
        s=r['href']
        t.append(s)

h=[]
for i in t:
    try:
        response=urllib.request.urlopen(i)
        html=response.read()
        soup=BeautifulSoup(html)

        text=soup.get_text()
        sentences= re.split(r' *[\.\?I][\'"\)\]]*', text)
        
        for s in sentences: s
            h.append(s)
        print(text)
                            
                            
                            
        

    except Exception as e:
        print(e, end="")
```

```python
h
```

```python
import pandas as pd
df=pd.DataFrame(h,columns=['Review'])
df
```

```python
import re

def cleanTxt(text):
    text=re.sub('@[A-Za-z0-9]+', '', text) #Removes mentions (e.g., @username).
    text=re.sub('#', '', text) #Removes the hashtag symbol (#) but keeps the word that followed it.
    text=re.sub('RT[\s]+', '', text) #Removes "RT" (retweet) followed by any whitespace.
    text=re.sub('https?:\/\/\S+', '', text) #Removes hyperlinks starting with http or https.
    text=re.sub('0', '', text) #Specifically removes the digit 0.
    text=re.sub('[0-9]', '', text) #Removes all digits from 0 to 9. (Note: The previous line is redundant because this covers all numbers).
    text=re.sub('%', '', text) #Removes percent signs.
    text=re.sub('\n', '', text) #Removes newline characters to make the text a single line.
    text=re.sub('\r', '', text)
    text=re.sub('\t', '', text)
    return text
    
```

```python
df['Review']=df['Review'].apply(cleanTxt)
```

### MOVIE REVIEWS USING AFINN(SHOW SCORE OF REVIEWS)

```python
from afinn import Afinn

positive=[]
negative=[]
neutral=[]

afinn=Afinn()

for review in df['Review']:
    if afinn.score(review) > 0:
        positive.append(review)

    elif afinn.score(review) < 0:
        negative.append(review)

    elif afinn.score(review) == 0:
        neutral.append(review)
```

```python
print ("Positive Reviews:", positive)

print('\t\n')

print("\nNegative Reviews:", negative)

print('\t\n')

print("\nNeutral Reviews:", neutral)

print('\t\n')

total_reviews = len(df['Review'])

print("Total number of reviews:", total_reviews)

print("Number of positive reviews:", len(positive))

print ("Number of negative reviews:", len(negative))

print("Numben of neutral reviews:", len(neutral))
```

### SENTIMENT ANALYSIS ON FLIPKART DATA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

data=pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/flipkart_reviews.csv")
print(data.head())
```

```python
import nltk
import re
nltk.download('stopwords')
from nltk.stem.snowball import SnowballStemmer
stemmer=SnowballStemmer('english')
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))



def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub('https?://\S+|www\.\s+','', text)

    text = re.sub('<.*?>+','', text)

    text = re.sub('[%s]' % re.escape(string.punctuation),'', text)

    text = re.sub('\n','', text)

    text = re.sub('\w*\d\w*', '', text)

    text = [word for word in text.split(' ') if word not in stopword]

    text=" ".join(text)

    text = [stemmer.stem(word) for word in text.split(' ')]

    text=" ".join(text)

    return text


data['Review']=data['Review'].apply(clean)

```

```python
def generate_context(row):
    # Extract the percentage value from the row
    seat_share = row["Seat_Share_%"]

    # Assign narrative based on percentage thresholds
    if seat_share >= 15:
        return f"secured a massive victory with {seat_share:.1f}% seat share."
    elif seat_share >= 2:
        return f"performed well, capturing a solid {seat_share:.1f}% seat share."
    else:
        return f"faced a tough election, winning only a minor {seat_share:.1f}% seat share."


# Apply the logic row-wise to create the new column
data1["Context_Sentence"] = data1.apply(generate_context, axis=1)
```
