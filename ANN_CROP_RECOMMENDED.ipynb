```python
# Encode Crop Labels
# --------------------------------------------------

from sklearn.preprocessing import LabelEncoder

# Initialize the encoder
le = LabelEncoder()

# Fit and transform the specific column
df['label'] = le.fit_transform(df['label'])

# # One-Hot Encoding
# y_categorical = to_categorical(y_encoded)
```

#### INPUT LAYER

```python
x = df.drop("label", axis=1)
y = df.iloc[0:, 7:8]
```

```python
x
```

```python
y
```

#### MODEL TRAIN

```python
# Train-Test Split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

#### feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

##### BUILD ANN MODEL AND HIDDEN LAYER(H1)

```python
# Build ANN Model and HIDDEN layers
# Index 0 (train_sc.shape[0]): Represents the total number of rows (sample size).
# Index 1 (train_sc.shape[1]): Represents the total number of columns (features or variables).

classifier = Sequential()

classifier.add(Dense(units=64, activation='relu',input_shape=(x_train_sc.shape[1],)))
classifier.add(Dropout(0.3))
```

#### H2

```python
classifier.add(Dense(units=64, activation='relu'))
classifier.add(Dropout(0.3))
```

#### H3

```python
classifier.add(Dense(units=32, activation='relu'))
```

#### ACTIVATION FUNCTION: SOFTMAX for MULTI-CATEGORY

```python
classifier.add(Dense(units=22,activation='softmax'))
```

#### OUTPUT LAYER

```python
# Compile Model

classifier.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train Model

classifier.fit(x_train_sc, y_train, batch_size=32, epochs=100)
```

```python
# Since the output is a set of probabilities that sum to 1.0 (if using Softmax), 
# you need to find the index of the highest value. This index corresponds to the crop the model "thinks" is the best fit for the given soil and 
# weather parameters.
# Math Logic: Predicted_Class = argmax(Output_Layer_Probabilities)



# Argmax (short for "argument of the maximum") is a function that tells you where the maximum value is, rather than what that value is.
# In machine learning, models often output a list of scores or probabilities for different categories (like "Cat," "Dog," or "Bird"). 
# While max would tell you the highest probability, argmax tells you which category that probability belongs to.


```

```python
# [0] means: вЂњGive me the first item from this array.вЂќ
# inverse_transform() в†’ returns a list/array of labels
# even if there is only one label, it still comes as a list
# [0] is used to get the actual string
```

```python
def recomand(a,b,c,d,e,f,g):
    input_data=np.array([[a,b,c,d,e,f,g]])
    scaled_input=sc.transform(input_data)
    pred=classifier.predict(scaled_input, verbose=True)
    pred_class_idx=np.argmax(pred)
    recomand_crop_name=le.inverse_transform([pred_class_idx])
    confidence=pred[0][pred_class_idx]*100
    return recomand_crop_name[0],confidence
    
```

```python
sample_crop,sample_conf=recomand(104,	18,	30,	23.603016,	60.396475,	6.779833,	140.937041)
sample_crop, sample_conf
```
