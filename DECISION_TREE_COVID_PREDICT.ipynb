```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\vv\OneDrive\Desktop\PYTHON\Covid Data.csv")
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
t=df.select_dtypes('int','float')
t
```

```python
t.corr()
```

```python
x = t[['USMER','INTUBED','COPD','ASTHMA','INMSUPR','HIPERTENSION','CARDIOVASCULAR','TOBACCO','ICU']]
x
```

```python
# classification: covid test findings. Values 1-3 mean that the patient was diagnosed with covid in different
# degrees. 4 or higher means that the patient is not a carrier of covid or that the test is inconclusive.
y=t.iloc[:,-2]
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=1000)
```

```python
from sklearn.tree import DecisionTreeClassifier

cls=DecisionTreeClassifier(criterion= 'entropy', random_state=1000)
cls.fit(x_train,y_train)
```

```python
y_pred=cls.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

results=confusion_matrix(y_test,y_pred)
print('confusion matrix')
print(results)
print('Accuracy score:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
