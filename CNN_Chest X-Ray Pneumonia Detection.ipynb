```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np
```

#### INITIATED THE CNN

```python
CNN_Classifier=Sequential()
```

```python
# step 1 : Convultion

CNN_Classifier.add(Conv2D(32,(3,3),input_shape=(64,64,3), activation='relu'))
```

```python
# step:2 Pooling

CNN_Classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
# step:1 Convulution

CNN_Classifier.add(Conv2D(32,(3,3), activation='relu'))
```

```python
# step:2 Pooling

CNN_Classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
# step 3: Flatten

CNN_Classifier.add(Flatten())
```

```python
# step:4 Full Connection

CNN_Classifier.add(Dense(units=128, activation='relu'))
```

```python
CNN_Classifier.add(Dense(units=1, activation='sigmoid'))
```

```python
CNN_Classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from keras_preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    shear_range=0.2,
    horizontal_flip=True
)

test_datagen= ImageDataGenerator(rescale= 1./255)

training_set=train_datagen.flow_from_directory(r"C:\Users\vv\Downloads\New folder\chest xray\train", target_size=(64,64),
                                               batch_size=16, class_mode='binary')

val_set = test_datagen.flow_from_directory(r"C:\Users\vv\Downloads\New folder\chest xray\val",
    target_size=(64,64),
    batch_size=16,
    class_mode='binary'
)

test_set = test_datagen.flow_from_directory(r"C:\Users\vv\Downloads\New folder\chest xray\test",target_size=(64,64),
                                            batch_size=16, class_mode='binary',shuffle=False)
                                            

CNN_Classifier.fit(training_set,
    validation_data=val_set,
    epochs=10
)
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np


img_path = r"C:\Users\vv\OneDrive\Desktop\PYTHON\DEEP LEARNING\Pneumonia_x_ray.jpg"

img = image.load_img(img_path,target_size=(64,64))

img = image.img_to_array(img)
img = img / 255.0

img = np.expand_dims(img, axis=0)

prediction = CNN_Classifier.predict(img)

probability = prediction[0][0]

print("Probability:", probability)

if probability > 0.5:
    print("PNEUMONIA DETECTED")
else:
    print("NORMAL")
```
